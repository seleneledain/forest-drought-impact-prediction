# Checking if NDVI is correct

```python
import os
import numpy as np
import rasterio
from rasterio.windows import from_bounds
import re
import matplotlib.pyplot as plt 
```

```python
# Load an image NDVI

ndvi_path = '/dccstor/cimf/drought_impact/downloads/SENTINEL 2/'

date = "2019-08-03T00:00:00Z_46.907_7.137_47.407_7.637/"
date = "2018-08-18T00:00:00Z_46.907_7.137_47.407_7.637/"
```

```python
mean_diff = []

for folder in os.listdir(ndvi_path):
    if '2018-08' in folder:
        print(folder)
        for file in os.listdir(ndvi_path+folder):
            if file.endswith('.tiff'):
                if bool(re.search('\[' +  'B08' + '\]',file)): 
                    print('NIR')
                    with rasterio.open(ndvi_path+folder+'/'+file) as blue_ds:
                        b8 = blue_ds.read(1,  masked=True)
                if bool(re.search('\[' +  'B04' + '\]',file)): 
                    print('Red')
                    with rasterio.open(ndvi_path+folder+'/'+file) as red_ds:
                        b4 = red_ds.read(1,  masked=True)
                if bool(re.search('\[' +  'NDVI' + '\]',file)): 
                    print('NDVI')
                    with rasterio.open(ndvi_path+folder+'/'+file) as ndvi_ds:
                        ndvi_band = ndvi_ds.read(1,  masked=True)
                        
    # Mean difference 
    NDVI_manual = (b8-b4)/(b8+b4)
    mean_diff += (ndvi_band-NDVI_manual).flatten().mean()
```

```python
mean_diff.mean()
```
